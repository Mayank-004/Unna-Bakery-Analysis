```python
for i in range(len(df_new1)):
    if pd.isnull(df_new1['Street'].iloc[i]) is True:
        r = usaddress.parse(df_new1['Address On File'].iloc[i])
        my_dict = dict(r)
        my_dict_new = {v:k for k,v in my_dict.items()}
        if ('AddressNumber' in my_dict_new) & ('StreetName' in my_dict_new)& ('StreetNamePostType' in my_dict_new):
            df_new1['Street'].iloc[i]= my_dict_new['AddressNumber'] +' '+  my_dict_new['StreetName'] + ' '+ my_dict_new['StreetNamePostType']
        else:
            df_new1['Street'].iloc[i] = 'undefined'
        if 'PlaceName' in my_dict_new:
            df_new1['City'].iloc[i] = my_dict_new['PlaceName']
        else:
            df_new1['City'].iloc[i] = 'undefined'
        if 'StateName' in my_dict_new:
            df_new1['State'].iloc[i]= my_dict_new['StateName']
        else:
            df_new1['State'].iloc[i]= 'undefined'
        if 'ZipCode' in my_dict_new:
            df_new1['Zip Code'].iloc[i] = my_dict_new['ZipCode']
        else:
            df_new1['Zip Code'].iloc[i] = 'undefined'
    else:
        continue
```
